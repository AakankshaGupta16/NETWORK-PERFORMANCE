```python
import speedtest
import pandas as pd
import time
import random

# File to store speed test logs
CSV_FILE = "/content/drive/MyDrive/network_speed_log.csv"

# Adjustable settings
INTERVAL = 60  # Interval between tests (in seconds)
TEST_COUNT = 6  # Number of tests to run

# Function to run a speed test
def run_speed_test():
    try:
        st = speedtest.Speedtest()
        st.get_servers([])
        st.get_best_server()  # Select the best available server

        download_speed = st.download() / 1e6  # Convert to Mbps
        upload_speed = st.upload() / 1e6  # Convert to Mbps
        ping = st.results.ping

        return download_speed, upload_speed, ping
    except speedtest.ConfigRetrievalError:
        print("⚠️ Speedtest configuration retrieval failed. Retrying in 30 seconds...")
        time.sleep(30)
        return run_speed_test()
    except Exception as e:
        print(f"❌ Unexpected error: {e}")
        return None, None, None

# Function to log speed test results in CSV
def log_speed_data():
    columns = ["Time", "Download Speed (Mbps)", "Upload Speed (Mbps)", "Ping (ms)"]

    try:
        df = pd.read_csv(CSV_FILE)
    except FileNotFoundError:
        df = pd.DataFrame(columns=columns)

    # Run speed test
    download_speed, upload_speed, ping = run_speed_test()

    if download_speed is None:
        print("❌ Skipping this test due to an error.")
        return

    # Create a new log entry
    new_entry = pd.DataFrame([[
        pd.Timestamp.now(), download_speed, upload_speed, ping
    ]], columns=columns)

    # Append to CSV
    df = pd.concat([df, new_entry], ignore_index=True)
    df.to_csv(CSV_FILE, index=False)

    # Print formatted log output
    print("\n📊 Speed Test Logged:")
    print(f"🕒 Time: {new_entry.iloc[0]['Time']}")
    print(f"⬇️ Download Speed: {new_entry.iloc[0]['Download Speed (Mbps)']:.2f} Mbps")
    print(f"⬆️ Upload Speed: {new_entry.iloc[0]['Upload Speed (Mbps)']:.2f} Mbps")
    print(f"⚡ Ping: {new_entry.iloc[0]['Ping (ms)']} ms")
    print("-" * 50)

# Run speed tests at intervals
print(f"🚀 Running {TEST_COUNT} speed tests every {INTERVAL} seconds...\n")

for i in range(TEST_COUNT):
    print(f"▶️ Running test {i+1}/{TEST_COUNT}...")
    log_speed_data()

    if i < TEST_COUNT - 1:
        delay = random.randint(INTERVAL, INTERVAL + 30)  # Random delay to avoid rate limits
        print(f"⏳ Waiting {delay} seconds before next test...")
        time.sleep(delay)

print("\n✅ Speed test logging completed.")
```

```python
import speedtest
import pandas as pd
import os

# File to store WiFi speed test logs
CSV_FILE = "/content/drive/MyDrive/wifi_data.csv"

# Function to run a speed test
def run_speed_test():
    try:
        st = speedtest.Speedtest()
        st.get_best_server()  # Select the best available server

        download_speed = st.download() / 1e6  # Convert to Mbps
        upload_speed = st.upload() / 1e6  # Convert to Mbps
        ping = st.results.ping

        print("\n📊 WiFi Speed Test Results:")
        print(f"⬇️ Download Speed: {download_speed:.2f} Mbps")
        print(f"⬆️ Upload Speed: {upload_speed:.2f} Mbps")
        print(f"⚡ Ping: {ping} ms")
        print("-" * 50)

        return download_speed, upload_speed, ping
    except speedtest.ConfigRetrievalError:
        print("⚠️ Speedtest configuration retrieval failed. Please try again later.")
        return None, None, None
    except Exception as e:
        print(f"❌ Unexpected error: {e}")
        return None, None, None

# Function to log speed test results in CSV
def log_speed_data():
    columns = ["Time", "Download Speed (Mbps)", "Upload Speed (Mbps)", "Ping (ms)"]

    try:
        df = pd.read_csv(CSV_FILE)
    except (FileNotFoundError, pd.errors.EmptyDataError):
        df = pd.DataFrame(columns=columns)

    # Run speed test
    download_speed, upload_speed, ping = run_speed_test()

    if download_speed is None:
        print("❌ Skipping this test due to an error.")
        return

    # Create a new log entry
    new_entry = pd.DataFrame([[
        pd.Timestamp.now(), download_speed, upload_speed, ping
    ]], columns=columns)

    # Append to CSV
    df = pd.concat([df, new_entry], ignore_index=True)
    df.to_csv(CSV_FILE, index=False)

    print("✅ Speed Test Logged Successfully.")

# Run a single speed test and log the results
print("🚀 Running WiFi Speed Test...")
log_speed_data()
print("✅ Speed Test Completed.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the network speed log CSV file
file_path = "/content/drive/MyDrive/network_speed_log.csv"
df = pd.read_csv(file_path)

# Convert Time column to datetime format
df['Time'] = pd.to_datetime(df['Time'])

# Set seaborn style
sns.set_style("whitegrid")

# Function to plot trends efficiently
def plot_trend(y_column, title, color):
    plt.figure(figsize=(8, 4))
    plt.plot(df['Time'], df[y_column], marker='o', linestyle='-', label=title, color=color)
    plt.xlabel("Time")
    plt.ylabel(y_column)
    plt.title(f"{title} Over Time")
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid()
    plt.show()

# Plot trends
plot_trend('Download Speed (Mbps)', "Download Speed", "b")
plot_trend('Upload Speed (Mbps)', "Upload Speed", "g")
plot_trend('Ping (ms)', "Ping", "r")

# Boxplot for Download and Upload Speeds
plt.figure(figsize=(6, 4))
sns.boxplot(data=df[['Download Speed (Mbps)', 'Upload Speed (Mbps)']])
plt.title("Network Speed Performance Distribution (Speed)")
plt.ylabel("Speed (Mbps)")
plt.grid()
plt.show()

# Separate boxplot for Ping
plt.figure(figsize=(6, 4))
sns.boxplot(data=df[['Ping (ms)']], color='r')
plt.title("Ping Performance Distribution")
plt.ylabel("Ping (ms)")
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cellular network speed log CSV file with encoding handling
file_path = "/content/drive/MyDrive/cellular_speed_log.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Print column names to verify correct headers
print("Column names in CSV:", df.columns.tolist())

# Attempt to find column names dynamically
column_mappings = {
    'Download Speed (Mbps)': ['download_speed', 'download mbps', 'dl_speed', 'down_speed'],
    'Upload Speed (Mbps)': ['upload_speed', 'upload mbps', 'ul_speed', 'up_speed'],
    'Ping (ms)': ['ping', 'latency', 'ping time']
}

# Standardize column names
for standard_col, possible_names in column_mappings.items():
    for name in possible_names:
        if name in df.columns:
            df.rename(columns={name: standard_col}, inplace=True)
            break

# Convert Time column to datetime format
df['Time'] = pd.to_datetime(df['Time'], errors='coerce')

# Set seaborn style
sns.set_style("whitegrid")

# Function to plot trends efficiently
def plot_trend(y_column, title, color):
    if y_column in df.columns:
        plt.figure(figsize=(8, 4))
        plt.plot(df['Time'], df[y_column], marker='o', linestyle='-', label=title, color=color)
        plt.xlabel("Time")
        plt.ylabel(y_column)
        plt.title(f"{title} Over Time")
        plt.legend()
        plt.xticks(rotation=45)
        plt.grid()
        plt.show()
    else:
        print(f"⚠️ Column '{y_column}' not found in the dataset.")

# Plot trends
plot_trend('Download Speed (Mbps)', "Download Speed", "b")
plot_trend('Upload Speed (Mbps)', "Upload Speed", "g")
plot_trend('Ping (ms)', "Ping", "r")

# Boxplot for Download and Upload Speeds
valid_columns = [col for col in ['Download Speed (Mbps)', 'Upload Speed (Mbps)'] if col in df.columns]
if valid_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df[valid_columns])
    plt.title("Cellular Network Speed Performance Distribution (Speed)")
    plt.ylabel("Speed (Mbps)")
    plt.grid()
    plt.show()
else:
    print("⚠️ No valid columns found for speed boxplot.")

# Separate boxplot for Ping
if 'Ping (ms)' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df[['Ping (ms)']], color='r')
    plt.title("Cellular Network Ping Performance Distribution")
    plt.ylabel("Ping (ms)")
    plt.grid()
    plt.show()
else:
    print("⚠️ 'Ping (ms)' column not found for ping boxplot.")
```
